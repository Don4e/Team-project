### MLP (Multilayer Perceptron) модел ###
Опис на моделот

MLP (Multilayer Perceptron) е невронска мрежа составена од повеќе слоеви на поврзани неврони, која овозможува моделирање на сложени и нелинеарни односи помеѓу влезните карактеристики и целната променлива. Во контекст на кредитна проценка, MLP е способен да препознае комплексни шаблони во податоците кои не можат лесно да се опфатат со линеарни модели, што го прави погоден за задачи каде интеракциите меѓу карактеристиките се значајни.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("../loan_dataset_20000.csv") 

X_raw = df.drop("loan_paid_back", axis=1)
y = df["loan_paid_back"]

# monthly and yearly income carry the same info monhly = yearly / 12
X_raw = X_raw.drop(columns=['monthly_income'])
```

### Pipeline и подготовка на податоците ###

MLP моделот беше имплементиран во рамки на pipeline, кој обезбедува конзистентна, репродуцибилна и безбедна обработка на податоците. Pipeline-от вклучуваше енкодирање на категоријалните карактеристики и скалирање на нумеричките податоци, по што трансформираните податоци директно се користеа за тренирање на моделот. Енкодирањето е неопходно бидејќи невронските мрежи работат исклучиво со нумерички вредности и не можат директно да обработуваат категоријални податоци. Скалирањето на нумеричките карактеристики е особено важно кај MLP моделите, бидејќи тие се тренираат со оптимизациски алгоритми базирани на градиент, при што разлики во размерот на влезните податоци можат да доведат до нестабилно или бавно учење и доминација на одредени карактеристики. Со стандардизација на податоците се обезбедува сите карактеристики да имаат споредлив опсег, што овозможува поефикасна и постабилна конвергенција на моделот. Дополнително, користењето pipeline гарантира дека истите трансформации се применуваат и при тренирање и при евалуација, со што се избегнува data leakage и се обезбедува коректна и реална проценка на перформансите на моделот.

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split


X = X_raw.copy()
y = y.copy()

categorical_cols = X.select_dtypes(include='object').columns.tolist()
numeric_cols = X.select_dtypes(include=['int64', 'float64']).columns.tolist()



preprocessor_scaled = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
    ]
)



target_col = "loan_paid_back" 
X = df.drop(columns=[target_col])
y = df[target_col]

# stratify keeps the 0/1 ratio similar in train and test (important for FP/FN work)
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Train shapes:", X_train.shape, y_train.shape)
print("Test shapes: ", X_test.shape, y_test.shape)
print("\nClass distribution (train):\n", y_train.value_counts(normalize=True))
print("\nClass distribution (test):\n", y_test.value_counts(normalize=True))
```

```python
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(
        hidden_layer_sizes=(32, 16),
        activation="relu",
        alpha=1e-3,
        max_iter=500,
        early_stopping=True,
        random_state=42
    )
pipeline_mlp = Pipeline([
    ('prep', preprocessor_scaled),
    ('model', mlp)
])

pipeline_mlp.fit(X_train, y_train)
```

```python
y_pred_mlp = pipeline_mlp.predict(X_test)
y_proba_mlp = pipeline_mlp.predict_proba(X_test)[:, 1]
```

### Резултати и евалуација ###

Евалуацијата на MLP моделот беше извршена со користење на конфузиона матрица и classification report, кои се доволни за детална анализа на перформансите кај бинарна класификација. Конфузионата матрица овозможува јасен увид во бројот на точни и погрешни предвидувања по класи, додека classification report ги прикажува клучните метрики како precision, recall и F1-score. Овие метрики овозможуваат балансирана проценка на моделот, особено во контекст на кредитен ризик каде различните типови на грешки имаат различна тежина. Резултатите покажуваат дека моделот е многу ефикасен во идентификација на клиенти кои ќе го вратат кредитот, но има простор за подобрување во детекцијата на ризични клиенти.

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred_mlp)

plt.figure(figsize=(5,4))
sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=['Predicted: Not Paid', 'Predicted: Paid'],
    yticklabels=['Actual: Not Paid', 'Actual: Paid']
)
plt.title('Confusion Matrix – MLP')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.tight_layout()
plt.show()
print(classification_report(y_test, y_pred_mlp))
```

### Анализа на важноста на карактеристиките ###

Бидејќи MLP моделот претставува невронска мрежа со нелинеарни трансформации, тој нема вградена и директна мерка за важност на карактеристиките. Затоа, влијанието на поединечните карактеристики беше анализирано со permutation importance, модел-агностичка техника која мери колкав пад во перформансите на моделот настанува при нарушување на вредностите на секоја карактеристика. На овој начин беше добиен рангиран список на највлијателни карактеристики, а дополнително беше прикажан и график за топ 10 карактеристики, што овозможува интерпретабилен и јасен увид во тоа кои фактори најмногу придонесуваат кон одлуките на MLP моделот.

```python
from sklearn.inspection import permutation_importance


# feature names after preprocessing
prep = pipeline_mlp.named_steps["prep"]
feature_names = prep.get_feature_names_out()

# compute permutation importance
result = permutation_importance(
    pipeline_mlp,
    X_test,
    y_test,
    n_repeats=10,
    random_state=42,
    scoring="f1"
)

# sort and print
importances = result.importances_mean
sorted_idx = np.argsort(importances)[::-1]

for idx in sorted_idx:
    print(f"{feature_names[idx]} : {importances[idx]}")
```

```python

# sort and take top 10
indices = np.argsort(importances)[-10:]
top_features = feature_names[indices]
top_importances = importances[indices]

# plot
plt.figure(figsize=(8, 5))
plt.barh(top_features, top_importances)
plt.xlabel("Permutation Importance (mean decrease in F1)")
plt.title("Top 10 Feature Importances – MLP")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
