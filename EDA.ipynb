```python
import pandas as pd
import numpy as np

df = pd.read_csv("loan_dataset_20000.csv") 
df.head()
```

```python
df.shape
df.info()
df['loan_paid_back'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='loan_paid_back', data=df)
plt.title('Distribution of Loan Repayment Status')
plt.xlabel('Loan Paid Back')
plt.ylabel('Count')
plt.show()
```

```python
df.columns
```

```python
fig, ax = plt.subplots(1,2, figsize=(12,4))
sns.histplot(df['credit_score'], kde=True, ax=ax[0])
sns.boxplot(x=df['credit_score'], ax=ax[1])
plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize=(12,4))
sns.histplot(df['debt_to_income_ratio'], kde=True, ax=ax[0])
sns.boxplot(x=df['debt_to_income_ratio'], ax=ax[1])
plt.show()
```

```python
sns.countplot(y='employment_status', data=df)
plt.title('Employment Status Distribution')
plt.show()
```

```python
sns.boxplot(x='loan_paid_back', y='credit_score', data=df)
plt.show()
```

```python
sns.boxplot(x='loan_paid_back', y='debt_to_income_ratio', data=df)
plt.show()
```

```python
corr = df.select_dtypes(include=['int64','float64']).corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr, cmap='coolwarm', center=0)
plt.show()
```
